```python
# no of cities to host summer olympics
#directly reading csv from link without downlaoding csv to local
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/AshishJangra27/Skill-India-AI-ML-Scholarship/main/Summer%20Olympics/summer.csv")
df.head()
len(df["City"].unique())
```

```python
# sport having most number of gold medals
from operator import itemgetter
l=[]
for i in df["Sport"].unique():
    l.append([i,len(df[(df["Sport"]==i) & (df["Medal"]=="Gold")])]) 
l.sort(key=itemgetter(1),reverse=True)
dta=pd.DataFrame(l,columns=["sport","count"]).head()
print(dta)
dta.plot(x="sport",y="count",kind="bar",figsize=(5,5))
```

```python
# sport having most number of medals
l=[]
for i in df["Sport"].unique():
    l.append([i,len(df[(df["Sport"]==i)])]) 
l.sort(key=itemgetter(1),reverse=True)
dta=pd.DataFrame(l,columns=["sport","count"]).head()
print(dta)
dta.plot(x="sport",y="count",kind="bar",figsize=(5,5))
```

```python
#player with most number of medals
l=[]
for i in df["Athlete"].unique():
    l.append([i,len(df[df["Athlete"]==i])]) 
l.sort(key=itemgetter(1),reverse=True)
dta=pd.DataFrame(l,columns=["Athlete","count"]).head()
print(dta)
dta.plot(x="Athlete",y="count",kind="bar",figsize=(5,5))
```

```python
#most popular Event with number of players
l=[]
for i in df["Event"].unique():
    l.append([i,len(df[(df["Event"]==i)])]) 
l.sort(key=itemgetter(1),reverse=True)
dta=pd.DataFrame(l,columns=["Event","count"]).head()
print(dta)
dta.plot(x="Event",y="count",kind="bar",figsize=(5,5))
```

```python
#sport with most female gold medalist
l=[]
for i in df["Sport"].unique():
    l.append([i,len(df[(df["Sport"]==i) & (df["Medal"]=="Gold")&(df["Gender"]=="Women")])]) 
l.sort(key=itemgetter(1),reverse=True)
dta=pd.DataFrame(l,columns=["sport","count"]).head()
print(dta)
dta.plot(x="sport",y="count",kind="bar",figsize=(5,5))
```
